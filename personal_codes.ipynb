#### Function to remove " - "

````python
removeHif = lambda x: { x: x.replace('-', '') }
````

#### Generate columns map

````python
# empty columns map
nameMap = {}

# populate columns map
for i in range(0, 100):
  nameWithG = f'g-{i}'
  nameWithC = f'c-{i}'
  nameMap.update(removeHif(nameWithG))
  nameMap.update(removeHif(nameWithC))
````

#### Replace columns
*_this will not replace original data, for replace original data we need to pass `inplace=True` as a param of `rename()`_

````python
dados.rename(columns=nameMap)
````

A função `loc()` permite acessar um elemento passando suas coordenadas.

Primeiro passamos o índice da linha, ou um range de indices usando `0:5` (da linha 0 até a 5) ou apenas `:` que vai pegar todas as linhas. Depois passamos a coluna, ou o range de colunas usando a mesma lógica das linhas.


_pega todas as linhas da coluna g-0 até a coluna g-771:_
```python

```

````python
# Compute the correlation matrix
corr = dados.loc[:, 'g-0':'g-50'].corr()
````

````python
import numpy as np
````

````python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
````

````python
# Compute the correlation matrix
corr_celular = dados.loc[:, 'c-0':'c-50'].corr()
````

````python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_celular, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_celular, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
````
